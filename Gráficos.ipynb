```python
import plotly.express as px
```

## Visão de Crescimento da Empresa

1- Quantidade de pedidos por dia, faça um Gráfico de barras.

```python
cols = ['ID','Order_Date']
df_aux = df.loc[:,cols].groupby('Order_Date').count().reset_index()

px.bar(df_aux, x='Order_Date', y='ID')
```

```python
df_aux.head(5)
```

2- Quantidade de pedidos por semana, faça um Gráfico de barras.

```python
df['Semanas_por_ano'] = df['Order_Date'].dt.strftime('%U')
df_aux = df.loc[:,['ID','Semanas_por_ano']].groupby('Semanas_por_ano').count().reset_index()
px.line(df_aux, x='Semanas_por_ano', y='ID')
```

```python
df_aux.head(5)
```

3- Distribuição dos pedidos por tipo de tráfego, faça um Gráfico de pizza

```python
df_aux = df.loc[:,['ID','Road_traffic_density']].groupby('Road_traffic_density').count().reset_index()
df_aux['Entregas_perc'] = df_aux['ID'] / df_aux['ID'].sum()
px.pie(df_aux, values = 'Entregas_perc', names='Road_traffic_density')
```

```python
df_aux.head(5)
```

4- Comparação do volume de pedidos por cidade e tipo de tráfego, façaa um Gráfico de barras

```python
df_aux = df.loc[:, ['ID','City','Road_traffic_density']].groupby(['City','Road_traffic_density']).count().reset_index()
px.scatter(df_aux,x='City', y='Road_traffic_density', size='ID', color='City')
```

```python
df_aux.head(5)
```

5- A quantidade de pedidos por entregador por semana, faça um Gráfico de linhas


```python
df_aux01 = df.loc[:,['ID','Semanas_por_ano']].groupby('Semanas_por_ano').count().reset_index()
df_aux02 = df.loc[:,['Delivery_person_ID','Semanas_por_ano']].groupby('Semanas_por_ano').nunique().reset_index()

df_aux= pd.merge(df_aux01,df_aux02, how='inner')

df_aux['Pedidos_entregador'] = df_aux['ID'] / df_aux['Delivery_person_ID']

px.line(df_aux, x='Semanas_por_ano', y='Pedidos_entregador')
```

```python
df_aux.head(5)
```

6- A localização central de cada cidade por tipo de tráfego, faça um Gráfico de Mapas


```python
import folium

df_aux = df.loc[:,['City','Road_traffic_density','Delivery_location_latitude','Delivery_location_longitude']].groupby(['City','Road_traffic_density']).median().reset_index()

map = folium.Map()

for index, location_info in df_aux.iterrows():
          folium.Marker ([location_info['Delivery_location_latitude'],
                          location_info['Delivery_location_longitude']]).add_to(map)

map

```
