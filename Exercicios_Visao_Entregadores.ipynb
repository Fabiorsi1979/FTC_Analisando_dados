```python
df = dados.copy()
```

# Visão dos Entregadores

### A menor e maior idade dos entregadores.

```python
df.loc[:,['Delivery_person_Age']].agg(['max','min'])
```

### A pior e a melhor condição de veículos.

```python
df.loc[:,['Vehicle_condition']].agg(['max','min'])
```

### A avaliação média por entregador.


```python
df.loc[:,['Delivery_person_ID','Delivery_person_Ratings']].groupby('Delivery_person_ID').mean().reset_index()
```

### A avaliação média e o desvio padrão por tipo de tráfego.

```python
df.loc[:,['Road_traffic_density','Delivery_person_Ratings']].groupby('Road_traffic_density').agg(['mean','std'])
```

### A avaliação média e o desvio padrão por condições climáticas.


```python
df.loc[:,['Weatherconditions','Delivery_person_Ratings']].groupby('Weatherconditions').agg(['mean','std'])
```

### Os 10 entregadores mais rápidos por cidade.


```python
cidades = df.City.unique()

df_aux = df.loc[:,['Delivery_person_ID','Time_taken(min)','City']].groupby(['City','Delivery_person_ID'])['Time_taken(min)'].mean().reset_index().sort_values(by='Time_taken(min)', ascending=False)

for cid in cidades:
    top_10 = df_aux[df_aux['City'] == cid].head(10)
    print(f"Top 10 entregadores na cidade {cid}:\n ", top_10)
```
